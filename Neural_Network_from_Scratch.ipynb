```python
inputs = [1, 2, 3, 2.5]
weights = [[0.2, 0.8, -0.5, 1],
[0.5, -0.91, 0.26, -0.5],
[-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

# Output of current layer
layer_outputs = []
# For each neuron
for neuron_weights, neuron_bias in zip(weights, biases):
# Zeroed output of given neuron
   neuron_output = 0
# For each input and weight to the neuron
   for n_input, weight in zip(inputs, neuron_weights):
# Multiply this input by associated weight
# and add to the neurons output variable
     neuron_output += n_input*weight
# Add bias
   neuron_output += neuron_bias
# Put neurons result to the layers output list
   layer_outputs.append(neuron_output)
print(layer_outputs)

```

Struktur dasar dari sebuah Artificial Neural Network (ANN), khususnya pada bagian Hidden Layer atau Dense Layer (lapisan padat/terhubung penuh).
1. Konsep Utama: Apa itu Dense Layer?Dalam sebuah Dense Layer, setiap neuron di lapisan tersebut menerima input dari semua neuron di lapisan sebelumnya. Secara matematis, setiap neuron melakukan operasi linear tunggal:$$y = \sum (inputs \times weights) + bias$$Atau dalam bentuk matriks untuk memproses banyak data sekaligus (batch):$$Y = X \cdot W + b$$

```python
import numpy as np


# Mengunci random seed agar hasil konsisten (menggantikan nnfs.init)

np.random.seed(0)

# Membuat dataset sederhana secara manual (menggantikan spiral_data)

# Misal: 3 sampel data, masing-masing memiliki 4 fitur

X = np.array([[1, 2, 3, 2.5],

[2.0, 5.0, -1.0, 2.0],

[-1.5, 2.7, 3.3, -0.8]])


class Layer_Dense:

  def __init__(self, n_inputs, n_neurons):

# Inisialisasi Weights: Matriks (Input x Neuron)

# Menggunakan 0.01 agar angka awal tidak terlalu besar/ekstrem

     self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)

# Inisialisasi Biases: Matriks (1 x Neuron)
# Dimulai dengan 0 karena bias berfungsi sebagai penggeser (offset)
     self.biases = np.zeros((1, n_neurons))



  def forward(self, inputs):

# Operasi utama: Output = (Inputs . Weights) + Biases

      self.output = np.dot(inputs, self.weights) + self.biases

# --- Eksekusi ---
# 1. Inisialisasi Layer: Input 4 (sesuai X), Neuron 3 (bebas)

layer1 = Layer_Dense(4, 3)

# 2. Jalankan Forward Pass

layer1.forward(X)

# 3. Lihat Hasilnya

print("Output Layer 1:\n", layer1.output)
```

Exercise: The Chain Reaction (Multi-Layer Network)

Deskripsi Tugas:  
Modifikasi kode yang ada untuk menciptakan sebuah jaringan saraf kecil dengan dua lapisan (Layer 1 dan Layer 2). Mahasiswa harus memastikan bahwa output dari Layer 1 dapat mengalir masuk ke Layer 2 tanpa menyebabkan dimension error.

Instruksi Latihan:  
1. Buat Layer ke-2: Inisialisasi objek baru bernama `layer2`.

2. Tentukan Dimensi: Berdasarkan kode asli, layer1 menghasilkan 3 neuron. Maka, [layer2](https://) harus menerima 3 input. Untuk jumlah neuron di `layer2`, tentukan sebanyak 2 neuron.

3. Hubungkan Alur (Forward Pass): * Jalankan forward pada layer1 menggunakan data X.

   Ambil hasil `layer1.output`, lalu gunakan itu sebagai input untuk fungsi forward pada `layer2`.

4. Tampilkan Hasil Akhir: Cetak output dari layer2.

```python
import numpy as np

# 1. Setup Lingkungan
np.random.seed(0)

# Input data (3 sampel, 4 fitur)
X = np.array([[1, 2, 3, 2.5],
              [2.0, 5.0, -1.0, 2.0],
              [-1.5, 2.7, 3.3, -0.8]])

# 2. Definisi Kelas (Tetap sama)
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        # Inisialisasi bobot dengan angka acak kecil
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        # Inisialisasi bias dengan nol
        self.biases = np.zeros((1, n_neurons))

    def forward(self, inputs):
        # Hitung dot product dan tambahkan bias
        self.output = np.dot(inputs, self.weights) + self.biases

# --- JAWABAN EXERCISE ---

# 3. Inisialisasi Dua Layer
# Layer 1: menerima 4 fitur, memprosesnya menjadi 3 neuron
layer1 = Layer_Dense(4, 3)

# Layer 2: HARUS menerima 3 input (karena Layer 1 punya 3 neuron)
# Kita ingin output akhirnya adalah 2 angka (misal untuk klasifikasi 2 kelas)
layer2 = Layer_Dense(3, 2)

# 4. Eksekusi Forward Pass Berantai (Forward Propagation)
layer1.forward(X)             # Langkah 1: X -> Layer 1
layer2.forward(layer1.output)  # Langkah 2: Output Layer 1 -> Layer 2

# 5. Menampilkan Hasil
print("--- Hasil Output Layer 1 (3 sampel, 3 neuron) ---")
print(layer1.output)
print("\n--- Hasil Output Layer 2 / Output Akhir (3 sampel, 2 neuron) ---")
print(layer2.output)
```
